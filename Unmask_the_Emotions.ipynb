```python
import pandas as pd
import numpy as np
import plotly.express as px
from textblob import TextBlob
```

```python
df = pd.read_csv('/content/netflix_titles.csv') # Create Data Frame and read the data from the csv file.
```

```python
df. shape # For the no. of Rows and Columns
```

```python
valid_ratings = ['TV-MA', 'PG-13', 'R', 'PG', 'G', 'NR', 'NC-17', 'TV-Y', 'TV-Y7', 'TV-G', 'NOT RATED']
df['rating'] = df['rating'].apply(lambda x: x if x in valid_ratings else 'OTHER')

# As Ratings column has the data which are not in rating terms as seen prima facie.
```

```python
df.head(5)
```

```python
df.columns
```

```python
x = df.groupby(['rating']).size().reset_index(name='counts')
print(x)
```

```python
piechart=px.pie(x,values='counts',names='rating',title='Distribution of Ratings')
piechart.show()
```

```python
df['director']=df['director'].fillna('No Director Specified')
df.head(3)
```

```python
director_list=pd.DataFrame()
print(director_list)
director_list['director']=df['director'].str.split(', ', expand=True).values.flatten()
print(director_list)

#This process extracts and flattens all individual director names from the original director column (which may contain multiple directors separated by commas) into a new DataFrame with one director name per row.
```

```python
director=director_list.groupby(['director']).size().reset_index(name='Total Counts')
print(director)
# Code calculates and displays the total number of shows or entries directed by each individual director in the dataset.
```

```python
director=director[director.director!='No Director Specified']
director=director.sort_values(by=['Total Counts'], ascending=False)
print(director)

# Code outputs a sorted and cleaned list of directors ranked by their frequency of appearances in the dataset.
```

```python
top_5_directors = director.head(5)  # Selects the first 5 rows after sorting

# Display the result
print(top_5_directors)

# Code extracts and prints the five most frequently occurring directors in the dataset based on the number of shows they directed
```

```python
fig = px.bar(top_5_directors, x='director', y='Total Counts',
             title='Top 5 Directors by Total Counts')

# Show the chart
fig.show()
```

```python
# 1. Clean and Prepare Cast Data

df['cast'] = df['cast'].fillna('No Cast Specified')  # Handle missing values
cast_list = pd.DataFrame()
cast_list['cast'] = df['cast'].str.split(', ', expand=True).values.flatten()
cast_counts = cast_list.groupby(['cast']).size().reset_index(name='Total Counts')
cast_counts = cast_counts[cast_counts.cast != 'No Cast Specified']  # Filter out 'No Cast Specified'
cast_counts = cast_counts.sort_values(by=['Total Counts'], ascending=False)  # Sort by counts

# 2. Get Top 5 Cast Members
top_5_cast = cast_counts.head(5)

# 3. Create Bar Chart
fig = px.bar(top_5_cast, x='cast', y='Total Counts',
             title='Top 5 Cast Members by Total Counts')

# 4. Show the Chart
fig.show()
```

```python
fig = px.histogram(df, x='release_year', title='Distribution of Release Years')
fig.show()

# Code generates an interactive histogram-highlighing the trends or peaks in content release overtime.
```

```python
fig = px.scatter(df, x='duration', y='rating',
                 title='Show Duration vs. Rating')
fig.show()
```

```python
genre_counts = df['listed_in'].str.split(', ', expand=True).stack().value_counts()
# Code extracts all individual genres from the multi-genre strings, then counts and sorts them by frequency, resulting in a Series showing the number of shows/movies for each genre in the dataset.
```

```python
fig = px.pie(values=genre_counts.values, names=genre_counts.index,
             title='Genre Proportions')
fig.show()
```

```python
titles_per_year = df.groupby('release_year').size().reset_index(name='counts')

# Code produces a DataFrame called titles_per_year that shows how many Netflix titles were released in each year, with two columns: release_year and the corresponding counts of titles.

fig = px.line(titles_per_year, x='release_year', y='counts',
              title='Titles Added to Netflix Over Time')
fig.show()
```

```python
# Maping the Sentiment from the description of the movies; Using TextBlob.
def get_sentiment(text):
    analysis = TextBlob(text)
    if analysis.sentiment.polarity > 0:
        return 'Positive'
    elif analysis.sentiment.polarity == 0:
        return 'Neutral'
    else:
        return 'Negative'

# Apply the function to the 'description' column
df['sentiment'] = df['description'].apply(get_sentiment)

# Explore the results
print(df[['description', 'sentiment']].head())  # Print some examples

```

```python
# Visualize sentiment distribution
sentiment_counts = df['sentiment'].value_counts()
fig = px.bar(x=sentiment_counts.index, y=sentiment_counts.values,
             title='Sentiment Distribution in Show Descriptions')
fig.show()
```
